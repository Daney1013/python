運算式

```python
#變數交換,透過第3變數z
x=1
y=10

z=x
x=y
y=z
print(x)
print(y)
```

```python
#變數交換, 不透過第3變數
x=1
y=10
x=x+y
y=x-y
x=x-y
print(x,y)
```

```python
#python版本的變數交換
x=1
y=10
x,y = y,x
print(x,y)
```

```python
#變數平行設定
x,y,z=1,10,100
print(x,y,z)
```

```python
x,y,z=3,4,5
x+=1 #x=x+1
y*=2 #y=y*2
z**=3 #z=z**3
print(x,y,z)
```

Q:半徑為10的圓形,其面積大小為何?

```python
import math
r=eval(input("請輸入圓的半徑:")) #eval()將字串轉化成數值運算
print(r*r*math.pi)
```
